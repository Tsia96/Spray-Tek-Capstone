### Creating new columns with date columns to convert the dates to hours for an effective Gantt Chart

```python
df3 = pd.read_csv('/Users/Admin/Desktop/MS BIA/Spring 2021/Industry Capstone Project/New DataFrames/SprayTek_Data.csv')
```

```python
df3.columns
```

```python
df3['dummyStartDate'] = df3['ActualStartDate'].iloc[0]
```

```python
df3['dummyEndDate'] = df3['Final_EndTime'].iloc[0]
```

```python
df3['OrderEntryDate'] = pd.to_datetime(df3['OrderEntryDate'])
df3['OrginalScheduleDate'] = pd.to_datetime(df3['OrginalScheduleDate'])
df3['ActualStartDate'] = pd.to_datetime(df3['ActualStartDate'])
df3['ReleaseDate'] = pd.to_datetime(df3['ReleaseDate'])
df3['Final_EndTime'] = pd.to_datetime(df3['Final_EndTime'])
df3['dummyStartDate'] = pd.to_datetime(df3['dummyStartDate'])
df3['dummyEndDate'] = pd.to_datetime(df3['dummyEndDate'])
```

```python
df3['StartingHour'] = df3['dummyStartDate'] - df3['ActualStartDate']
df3['EndingHour'] = df3['dummyEndDate'] - df3['Final_EndTime']
```

```python
df3['StartingHour'] = pd.TimedeltaIndex(df3['StartingHour'], unit = 'H')
```
